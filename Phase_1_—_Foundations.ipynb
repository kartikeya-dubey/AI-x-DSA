# 📘 Topic 1: Two Pointers & Sliding Window

## 🧠 The Theory

An agent's **working memory** at any given moment is its **context window**—the maximum number of tokens that a Large Language Model (LLM) can process in a single request.

When working with large documents, it is **impossible** to send the entire document to the model if it exceeds the context limit. Instead, the document must be divided into **smaller overlapping chunks**.

This is where the **Sliding Window** technique becomes essential.

The Sliding Window algorithm uses **two pointers** (indices representing the beginning and end of the current window) to process a continuous sequence of data efficiently. After processing one chunk, the window **slides forward**, usually with a small overlap, ensuring that important information spanning two chunks is not lost.

### Why is this important in Agentic AI?

- 📄 Text chunking for Retrieval-Augmented Generation (RAG)
- 🧩 Managing LLM token limits
- 🔄 Maintaining context across long conversations
- 📚 Document preprocessing before embedding
- ⚡ Efficient sequential data processing

---

## 💻 The Code (Run this in Google Colab)

The following example demonstrates how to split a long document into **overlapping chunks** using a Sliding Window.

The overlap preserves context between consecutive chunks so the agent can maintain continuity while processing the document.


```python
def sliding_window_chunker(text, window_size, overlap):
    """
    Simulates chunking a document for an LLM context window.
    """
    # In a real AI app, this would be a list of tokens.
    # For simplicity, we split by words.
    words = text.split()
    chunks = []

    # 'start' is our first pointer.
    # It moves forward by (window_size - overlap) each time.
    start = 0

    while start < len(words):
        # 'end' is our second pointer, defining the window's boundary.
        end = start + window_size

        # Extract the chunk using slicing (which operates via pointers under the hood)
        chunk = words[start:end]
        chunks.append(" ".join(chunk))

        # Slide the window forward, keeping the overlap
        start += (window_size - overlap)

    return chunks

# --- Test it out ---
document = """This document presents a comprehensive review of the regulatory tax reporting framework used by T-Files to support tax reporting obligations for banking sector clients operating across multiple jurisdictions. The review was conducted as part of the ongoing Regulatory Filings engagement delivered under the Tax service line, with the objective of evaluating the robustness, consistency, and regulatory alignment of tax reporting processes supporting financial regulatory submissions. T-Files provides technology-enabled reporting infrastructure used by financial institutions to prepare regulatory submissions required by supervisory authorities. These submissions often include disclosures related to corporate income tax reporting, withholding tax reporting, tax expense disclosures, and regulatory transparency obligations associated with financial reporting frameworks. Given the highly regulated nature of the banking sector, tax reporting processes must demonstrate a high degree of accuracy, documentation integrity, and audit traceability. The consulting team initiated this review following preliminary observations indicating that certain tax reporting activities relied on manual reconciliation processes and fragmented documentation practices. The objective of the review was therefore to assess how tax data flows from underlying financial systems into regulatory reporting outputs and to evaluate the adequacy of governance controls supporting these processes.
The review involved detailed analysis of reporting workflows used to generate tax disclosures included in regulatory filings submitted to financial supervisory authorities. These workflows involve the extraction of financial data from multiple transaction systems, transformation of that data into reporting formats required by regulatory templates, and final validation of tax figures before submission to regulatory bodies.
As part of the review process, the consulting team conducted structured interviews with key stakeholders involved in the tax reporting process, including finance leaders responsible for tax compliance, regulatory reporting specialists responsible for preparing supervisory filings, and IT personnel responsible for maintaining reporting infrastructure. These discussions provided valuable insight into how tax reporting processes operate in practice and revealed areas where operational complexity has increased due to evolving regulatory expectations.
One of the primary findings of the review relates to the complexity of mapping financial accounting data to tax reporting categories required for regulatory disclosures. Financial reporting systems often capture accounting classifications that do not directly align with tax reporting categories required by supervisory authorities. As a result, finance teams frequently perform manual adjustments to reconcile accounting data with tax disclosure requirements before regulatory reports can be finalized.
While these reconciliation procedures allow the organization to produce compliant regulatory reports, they also introduce operational risks. Manual adjustments increase the possibility of inconsistencies between underlying financial records and regulatory tax disclosures. Additionally, reliance on spreadsheet-based reconciliation processes reduces the level of system control and traceability associated with tax reporting workflows.
The consulting team also evaluated documentation practices supporting tax calculations used in regulatory filings. In several cases, supporting documentation was maintained in decentralized locations across multiple reporting teams. This approach made it difficult to establish a clear audit trail demonstrating how tax figures were derived and validated before submission to regulatory authorities.
From a regulatory compliance perspective, maintaining a clear and well-documented audit trail is particularly important in the banking sector. Financial supervisory authorities frequently conduct regulatory inspections to verify the accuracy of financial disclosures submitted by regulated institutions. During such inspections, institutions must be able to demonstrate how reported tax figures were calculated and how those calculations were validated through internal control procedures.
The review therefore emphasized the importance of establishing standardized documentation procedures for tax reporting processes. Centralized documentation repositories and standardized reporting templates can help ensure that tax calculations supporting regulatory filings are consistently documented and easily accessible during regulatory reviews.
Another area examined during the review relates to the governance framework overseeing tax reporting processes. The consulting team observed that responsibilities for tax reporting activities were distributed across several operational teams, including finance, regulatory reporting, and technology functions. While this structure allowed specialized teams to contribute expertise in their respective areas, it also created potential challenges in maintaining clear accountability for tax reporting outcomes.
To address this issue, the consulting team recommended strengthening governance structures responsible for overseeing tax reporting activities. Establishing clearly defined roles and responsibilities for tax reporting oversight would help ensure that regulatory reporting obligations are managed consistently across the organization.
In addition to governance considerations, the review also evaluated opportunities to improve automation within tax reporting workflows. Many of the manual reconciliation steps identified during the assessment could potentially be automated through enhancements to the reporting platform used by T-Files. Automating data mapping between financial accounting classifications and tax reporting categories would significantly reduce operational complexity and improve the reliability of regulatory tax disclosures.
Technology improvements could also enable the implementation of validation controls within reporting workflows. Such controls would allow reporting systems to automatically identify discrepancies between financial accounting data and tax reporting outputs before regulatory filings are finalized. These automated validation mechanisms would provide an additional layer of assurance that tax disclosures included in regulatory submissions are accurate and consistent.
Another important consideration highlighted during the review relates to the evolving regulatory landscape governing financial institutions. Tax reporting requirements for banking institutions continue to evolve as regulatory authorities introduce new transparency initiatives and reporting standards. As a result, reporting frameworks must be sufficiently flexible to accommodate future regulatory changes without requiring extensive manual reconfiguration.
The consulting team therefore emphasized the importance of designing tax reporting frameworks that are both robust and adaptable. Implementing modular reporting architectures and standardized data mapping structures can enable organizations to adapt more easily to changing regulatory requirements while maintaining consistent reporting practices.
Throughout the review process, the consulting team also considered how improvements to tax reporting processes could contribute to broader organizational objectives related to financial governance and risk management. Strengthening tax reporting controls not only supports regulatory compliance but also enhances the reliability of financial data used for internal decision-making and external reporting.
By improving transparency and documentation of tax reporting processes, organizations can reduce the likelihood of regulatory disputes and demonstrate a strong commitment to financial governance. This is particularly important in the banking sector, where regulatory expectations regarding financial transparency and reporting accuracy are exceptionally high.
The findings presented in this document therefore serve as the foundation for developing targeted recommendations aimed at strengthening T-Files’ regulatory tax reporting framework. These recommendations will focus on improving governance structures, enhancing documentation practices, and implementing technology-enabled controls designed to support consistent and reliable tax reporting processes.
The consulting team will work closely with finance, regulatory reporting, and technology stakeholders within T-Files to translate these recommendations into practical implementation plans. These plans will prioritize initiatives that deliver immediate improvements in reporting transparency while also establishing the foundation for long-term enhancements to regulatory tax reporting capabilities.
Ultimately, the objective of this engagement is to help T-Files strengthen its ability to support regulatory tax reporting obligations for banking sector clients, ensuring that financial institutions relying on the platform can meet their reporting obligations with confidence while maintaining compliance with evolving regulatory requirements."""

# We want chunks of 40 words, with 10 word overlapping to maintain context.
agent_chunks = sliding_window_chunker(document, window_size=40, overlap=10)

for i, chunk in enumerate(agent_chunks):
    print(f"Chunk {i+1}: {chunk}")
```
