# 📘 Topic 1: Priority Queues & Heaps (Task Schedulers)

---

# Layer 1 — Pure DSA Basics

## 🧠 The Theory

A **Heap** is a specialized **tree-based** data structure that satisfies the **Heap Property**.

The most commonly used variant is the **Min-Heap**, where the smallest element is always stored at the root.

```text
          2
        /   \
       5     7
      / \   / \
     9 10 12 15
```

Every parent node is **less than or equal to** its children.

Because of this property:

- The smallest element can always be accessed instantly.
- Insertion and deletion remain highly efficient.

---

## Why Not Just Sort an Array?

Imagine continuously receiving new tasks.

With a normal array:

1. Insert a task.
2. Sort the entire array.
3. Remove the highest-priority task.

Sorting repeatedly becomes expensive.

Time Complexity:

```text
O(N log N)
```

for every scheduling operation.

---

A **Heap** avoids repeated sorting.

Instead:

- Insert the new task while maintaining the heap property.
- Remove only the root when the next task is needed.

Time Complexity:

| Operation | Complexity |
|-----------|-----------:|
| Peek Minimum | **O(1)** |
| Insert | **O(log N)** |
| Extract Minimum | **O(log N)** |

This makes heaps the standard data structure for **priority scheduling**.

---

## Min-Heap vs Max-Heap

### Min-Heap

```text
        1
      /   \
     3     5
    / \   / \
   7  9  8 10
```

Always removes the **smallest** value first.

---

### Max-Heap

```text
        10
      /    \
     8      7
    / \    / \
   5   4  3   2
```

Always removes the **largest** value first.

Python's built-in `heapq` module implements a **Min-Heap**.

---

## Why Are Heaps So Fast?

Unlike a fully sorted array, a heap only guarantees one thing:

> **The root node is always the highest-priority element.**

The remaining elements only satisfy the local parent-child ordering.

This relaxed ordering allows insertion and deletion to take only:

```text
O(log N)
```

instead of sorting the entire collection after every update.

---

## 💻 Basic Python Code (Run in Google Colab)

In the following example, we'll use Python's built-in **`heapq`** module to create a **Min-Heap**, insert multiple values, and repeatedly extract the smallest element.

This demonstrates the core mechanics of a Priority Queue, which forms the foundation for task scheduling in modern Agentic AI systems.

````
import heapq
import random

# A basic list of numbers we want to manage by priority (smallest first)
raw_tasks = [40, 10, 30, 20, 50]
min_heap = []

# Inserting items into a heap (Heapify / Push)
for task in raw_tasks:
    heapq.heappush(min_heap, task)

print("Min-Heap Structure:", min_heap) # The smallest element (10) is always at index 0

i=1

# Extracting elements in order of priority (smallest first)
while min_heap:
    next_item = heapq.heappop(min_heap)
    print(f"Extracted highest priority item: {next_item}")
    if i<5:
      n=random.randint(1, 100)
      print(f"Inserting new item: {n}")
      heapq.heappush(min_heap,n)
    print("Updated Min-Heap Structure:", min_heap)
    i+=1
````

## Deterministic Scheduling

Python's built-in **`heapq`** module compares tuple elements **from left to right**.

A common production pattern is:

```python
(priority_score, timestamp, task_object)
```

Comparison order:

1. **Priority**
2. **Timestamp**
3. **Task object**

Example:

```python
(1, 1001, Task A)

(1, 1002, Task B)
```

Both tasks have the same priority.

Since the priorities are equal, the scheduler compares the timestamps.

```text
1001 < 1002
```

Therefore:

```text
Task A executes first.
```

This guarantees **deterministic execution**, ensuring that tasks with identical priorities are processed in the exact order they were created.

---

## Why This Matters in Enterprise Systems

Real-world schedulers must support:

- 📋 Thousands of concurrent tasks
- 🤖 Multiple autonomous agents
- 🔄 Automatic retries
- ⚡ Parallel execution
- 📈 Monitoring and auditing
- 🛡️ Fault tolerance

A priority queue enables the scheduler to efficiently determine the **next best task** without repeatedly sorting the entire task list.

Each scheduling operation requires only:

```text
O(log N)
```

making it highly scalable for production workloads.

---

## 💻 Production Code (Run in Google Colab)

In the following implementation, we'll build a simplified **Enterprise Task Scheduler** using Python's **`heapq`** module.

Each task will contain rich metadata—including a priority score, timestamp, unique task ID, execution status, and retry count—and the scheduler will always execute the highest-priority task first while maintaining deterministic ordering for tasks with equal priority.

````
import heapq
import time
import uuid
from typing import Dict, Any, Optional

class AgentTask:
    """
    Enterprise metadata structure for an agent sub-task.
    """
    def __init__(self, name: str, payload: Dict[str, Any], priority: int, max_retries: int = 3):
        self.task_id: str = f"task_{uuid.uuid4().hex[:8]}"
        self.name: str = name
        self.payload: Dict[str, Any] = payload
        self.priority: int = priority  # Lower number = Higher Priority (Min-Heap)
        self.max_retries: int = max_retries
        self.retry_count: int = 0
        self.status: str = "PENDING"  # PENDING, RUNNING, FAILED, COMPLETED
        self.created_at: float = time.time()

    def __lt__(self, other):
        """
        Custom comparator for Python's heapq.
        If priorities are equal, use timestamp (FIFO tie-breaker).
        """
        if self.priority == other.priority:
            return self.created_at < other.created_at
        return self.priority < other.priority

class EnterpriseTaskScheduler:
    """
    Priority Queue Task Scheduler for Autonomous Multi-Agent Workflows.
    """
    def __init__(self):
        self.task_heap: List[AgentTask] = []
        self.task_registry: Dict[str, AgentTask] = {} # Hash map for O(1) state lookup

    def enqueue_task(self, name: str, payload: Dict[str, Any], priority: int) -> AgentTask:
        task = AgentTask(name=name, payload=payload, priority=priority)
        self.task_registry[task.task_id] = task
        heapq.heappush(self.task_heap, task)
        print(f"[ENQUEUED] Task '{task.name}' (ID: {task.task_id}) with Priority Level {priority}")
        return task

    def get_next_critical_task(self) -> Optional[AgentTask]:
        """Pops the highest priority task in O(log N) time."""
        if not self.task_heap:
            return None
        task = heapq.heappop(self.task_heap)
        task.status = "RUNNING"
        return task

    def update_task_status(self, task_id: str, new_status: str, error_encountered: bool = False):
        """O(1) state update using registry hash map, with automatic failure recovery logic."""
        if task_id not in self.task_registry:
            raise KeyError(f"Task ID {task_id} not found.")

        task = self.task_registry[task_id]
        task.status = new_status

        if error_encountered and task.retry_count < task.max_retries:
            task.retry_count += 1
            task.status = "PENDING"
            # Lower priority slightly on retry by increasing priority integer value, then push back
            task.priority += 1
            heapq.heappush(self.task_heap, task)
            print(f" [RETRY WARNING] Task '{task.name}' failed. Re-queued (Attempt {task.retry_count}/{task.max_retries}) with adjusted priority {task.priority}.")
        else:
            print(f" [STATUS UPDATE] Task '{task.name}' -> State: {task.status}")

# =====================================================================
# LIVE SIMULATION IN COLAB
# =====================================================================
scheduler = EnterpriseTaskScheduler()

# 1. Agent generates a set of sub-tasks with varying urgency
scheduler.enqueue_task("Generate Final Executive Summary Report", {"format": "pdf"}, priority=3)
scheduler.enqueue_task("CRITICAL: Patch SQL Injection Vulnerability", {"cve": "CVE-2026-001"}, priority=1)
scheduler.enqueue_task("Fetch User Analytics Data from Warehouse", {"limit": 1000}, priority=2)

print("\n--- BEGINNING AGENT EXECUTION LOOP ---\nwhile tasks remain:")

# 2. Agent processes tasks strictly by priority
while scheduler.task_heap:
    current_task = scheduler.get_next_critical_task()
    print(f"Executing -> [{current_task.name}] (Priority: {current_task.priority})")

    # Simulate execution: Let's pretend the analytics task fails on first run
    if "Analytics" in current_task.name and current_task.retry_count == 0:
        scheduler.update_task_status(current_task.task_id, "FAILED", error_encountered=True)
    else:
        scheduler.update_task_status(current_task.task_id, "COMPLETED")
    print("-" * 50)
````

# 📘 Topic 2: Stacks (Recursive Agent Calls & Nested Tool-Use Evaluation)

---

# Layer 1 — Pure DSA Basics

## 🧠 The Theory

A **Stack** is a linear data structure that follows the **LIFO (Last-In, First-Out)** principle.

Think of a stack of plates.

You can only:

- Place a new plate on the **top**
- Remove the plate from the **top**

```text
Push

    ┌─────┐
    │  C  │  ← Top
    ├─────┤
    │  B  │
    ├─────┤
    │  A  │
    └─────┘
```

When removing elements:

```text
Pop

C ← removed first
B ← removed second
A ← removed last
```

The **last element inserted** is always the **first one removed**.

---

## Core Operations

### Push

Adds a new element to the top of the stack.

```text
Before

Top
 │
 ▼
[A]
[B]

Push(C)

Top
 │
 ▼
[C]
[A]
[B]
```

Time Complexity:

```text
O(1)
```

---

### Pop

Removes the top element.

```text
Before

Top
 │
 ▼
[C]
[A]
[B]

Pop()

Top
 │
 ▼
[A]
[B]
```

Time Complexity:

```text
O(1)
```

---

### Peek

Returns the top element without removing it.

```text
Top → C
```

Time Complexity:

```text
O(1)
```

---

### Is Empty

Checks whether the stack contains any elements.

```python
len(stack) == 0
```

Time Complexity:

```text
O(1)
```

---

## Why Are Stacks Efficient?

Unlike arrays, stacks only allow insertion and deletion from **one end**.

Because there is no need to shift existing elements, the primary operations remain constant time.

| Operation | Time Complexity |
|-----------|----------------:|
| Push | **O(1)** |
| Pop | **O(1)** |
| Peek | **O(1)** |
| Is Empty | **O(1)** |

---

## Common Applications of Stacks

Stacks are widely used for:

- 📞 Function call management
- 🔁 Recursion
- ↩️ Backtracking
- 🧮 Expression evaluation
- 🌐 Browser history
- ↶ Undo / Redo functionality
- 📂 Directory navigation

Whenever execution must **return to a previous state**, a stack is usually involved.

---

## 💻 Basic Python Code (Run in Google Colab)

In the following example, we'll implement a simple **Stack** using Python lists and demonstrate the core operations: **push**, **pop**, **peek**, and **is_empty**. These operations form the foundation for recursive algorithms, function call stacks, and the execution flow used in modern Agentic AI systems.

````
# A simple list acting as a stack
stack = []

# Push elements onto the stack
stack.append("Task_1")
stack.append("Task_2")
stack.append("Task_3")
print("Stack after pushes:", stack)

# Pop elements (The last one added comes off first)
top_item = stack.pop()
print(f"Popped item: {top_item}")
print("Stack after pop:", stack)
````

# Layer 2 — AI-Specific Context (Nested Tool Calls & Agent Sub-tasks)

## 🧠 The Theory

Modern AI agents rarely execute a single tool in isolation.

Instead, one tool frequently triggers another, creating a chain of **nested tool calls**.

Consider the following example.

An agent receives the request:

> **"Find the latest sales report and summarize it."**

The execution might look like this:

```text
Agent
 │
 ▼
Tool A: Search Web
 │
 ▼
Tool B: Query SQL Database
 │
 ▼
Tool C: Parse JSON Response
```

Notice that:

- **Tool A** cannot finish until **Tool B** completes.
- **Tool B** cannot finish until **Tool C** completes.

Execution therefore follows a **Last-In, First-Out (LIFO)** order.

The most recently invoked tool must complete before its parent can continue.

```text
Execution

Search Web
    │
    ▼
Query Database
    │
    ▼
Parse JSON

Return Order

Parse JSON
      ▲
Query Database
      ▲
Search Web
```

This behavior naturally maps to a **Stack**.

Every time a tool invokes another tool:

- Push the new execution frame onto the stack.

When the nested task finishes:

- Pop it from the stack.
- Resume the previous tool.

This is identical to how programming languages manage **function calls** internally using the **call stack**.

---

## 🤖 Why Stacks Matter in Agentic AI

Stacks enable agents to correctly manage:

- 🔧 Nested tool invocations
- 🔁 Recursive reasoning
- 📞 Function execution
- 🌳 Planning trees
- ↩️ Backtracking after failures
- 🧠 Chain-of-thought execution
- 🔍 Multi-step reasoning

Without a stack, an agent would quickly lose track of **which tool should resume execution** after a nested operation completes.

## Preventing Infinite Agent Loops

Autonomous agents can accidentally enter recursive execution loops.

Example:

```text
Tool A
   ↓
Tool B
   ↓
Tool C
   ↓
Tool A
```

Without safeguards, this cycle could continue indefinitely.

Enterprise orchestrators use stack metadata to enforce safety mechanisms such as:

- Maximum recursion depth
- Maximum tool-call count
- Execution time limits
- Retry thresholds
- Loop detection

If any limit is exceeded, the framework terminates the workflow or requests human intervention.

- 📈 Observability and debugging

Since **push** and **pop** operations both execute in **O(1)** time, stacks provide an efficient foundation for managing complex agent workflows without introducing unnecessary overhead.

## 💻 Production Code (Run in Google Colab)

In the following implementation, we'll build a simplified **Enterprise Tool Execution Stack** that tracks nested tool calls using rich execution metadata, including tool IDs, parent-child relationships, execution depth, timestamps, and status. The stack will simulate how production agent frameworks manage recursive workflows, maintain execution traces, and safely return control as nested tool calls complete.

````
import time
import uuid
from typing import Dict, Any, List, Optional

class ExecutionFrame:
    """
    Enterprise Call Stack Frame for tracking nested agent tool-use evaluations.
    """
    def __init__(self, tool_name: str, arguments: Dict[str, Any], depth: int):
        self.frame_id: str = f"frame_{uuid.uuid4().hex[:8]}"
        self.tool_name: str = tool_name
        self.arguments: Dict[str, Any] = arguments
        self.depth: int = depth
        self.start_time: float = time.time()
        self.status: str = "RUNNING" # RUNNING, COMPLETED, FAILED
        self.result: Optional[Any] = None

class AgentCallStack:
    """
    Enterprise LIFO Stack Manager for Recursive Agent Actions and Sub-tool Evaluations.
    """
    def __init__(self, max_recursion_depth: int = 5):
        self.stack: List[ExecutionFrame] = []
        self.max_recursion_depth: int = max_recursion_depth
        self.audit_log: List[ExecutionFrame] = [] # Keeps historical trace

    def push_call(self, tool_name: str, arguments: Dict[str, Any]) -> ExecutionFrame:
        current_depth = len(self.stack) + 1

        # Enterprise Safety Guard: Prevent Infinite Agent Loops
        if current_depth > self.max_recursion_depth:
            raise RecursionError(f"Agent exceeded maximum allowed recursion depth ({self.max_recursion_depth}). Infinite loop detected!")

        frame = ExecutionFrame(tool_name=tool_name, arguments=arguments, depth=current_depth)
        self.stack.append(frame)
        print(f"{'  ' * (current_depth - 1)}[PUSH] Entering tool '{tool_name}' (Depth: {current_depth})")
        return frame

    def pop_call(self, result: Any, status: str = "COMPLETED") -> ExecutionFrame:
        if not self.stack:
            raise IndexError("Stack underflow! No active tool calls to pop.")

        frame = self.stack.pop()
        frame.result = result
        frame.status = status
        elapsed_ms = (time.time() - frame.start_time) * 1000

        self.audit_log.append(frame)
        print(f"{'  ' * (frame.depth - 1)}[POP] Finished tool '{frame.tool_name}' in {elapsed_ms:.2f}ms -> Status: {status}")
        return frame

# =====================================================================
# LIVE SIMULATION IN COLAB
# =====================================================================
agent_stack = AgentCallStack(max_recursion_depth=4)

try:
    # Agent decides to call Tool A (Search)
    agent_stack.push_call("WebSearchTool", {"query": "latest Q1 financial metrics"})

    # Inside WebSearchTool, it triggers a nested Tool B (Database Lookup)
    agent_stack.push_call("SQLDatabaseTool", {"query": "SELECT revenue FROM q1_metrics"})

    # Inside SQLDatabaseTool, it triggers a deeper Tool C (JSON Parser / Formatter)
    agent_stack.push_call("DataParserTool", {"format": "normalize_decimal"})

    # Tool C finishes execution and pops off stack
    agent_stack.pop_call(result={"status": "normalized", "precision": 2})

    # Tool B finishes execution and pops off stack
    agent_stack.pop_call(result={"revenue": "$4.2M"})

    # Tool A finishes execution and pops off stack
    agent_stack.pop_call(result={"summary": "Q1 Revenue is $4.2M based on verified DB records."})

    print("\n--- AUDIT TRACE LOG ---")
    for logged_frame in agent_stack.audit_log:
        print(f"Tool: {logged_frame.tool_name} | Depth: {logged_frame.depth} | Result: {logged_frame.result}")

except RecursionError as e:
    print(f"Safety Triggered: {e}")
````

# 📘 Topic 3: Directed Acyclic Graphs (DAGs) for Agent Workflows)

---

# Layer 1 — Pure DSA Basics

## 🧠 The Theory

A **Graph** is a non-linear data structure consisting of:

- **Nodes (Vertices)** → Represent entities or states.
- **Edges** → Represent relationships or connections between nodes.

Unlike trees, graphs allow a node to connect with multiple other nodes, making them ideal for representing complex workflows and dependencies.

```text
      A
     / \
    B   C
     \ /
      D
```

---

## Directed Graph

In a **Directed Graph**, every edge has a specific direction.

```text
A ───► B
│
▼
C ───► D
```

Here:

- A can reach B.
- B cannot automatically return to A.
- Every connection represents a one-way relationship.

Directed graphs are commonly used to model:

- Task dependencies
- Workflow execution
- State transitions
- Network routing

---

## Directed Acyclic Graph (DAG)

A **Directed Acyclic Graph (DAG)** is a directed graph that contains **no cycles**.

In other words:

> Once you move forward through the graph, you can never return to a previously visited node.

Example:

```text
Start
  │
  ▼
Load Data
  │
  ▼
Analyze
 ├──────► Generate Report
 │
 ▼
Store Results
```

This is a DAG because there is **no path that leads back to an earlier node**.

---

### What is a Cycle?

A cycle occurs when a graph allows you to return to the starting node.

Example:

```text
A ─► B
▲    │
│    ▼
└────C
```

Here:

```text
A → B → C → A
```

Since the graph loops back to **A**, it is **not** a DAG.

Cycles can lead to:

- Infinite loops
- Deadlocks
- Endless recursion
- Repeated execution

For workflow engines, these are highly undesirable unless explicitly managed.

---

## Why Are DAGs Important?

DAGs guarantee that execution always moves **forward**.

This property makes them ideal for representing systems where tasks have dependencies.

Examples include:

- Build systems
- Data pipelines
- Workflow engines
- Scheduling systems
- Agent orchestration frameworks

Each task executes only after all of its required predecessors have completed.

---

## Graph Representation

One of the most common ways to represent a graph in Python is using an **Adjacency List**.

Example:

```python
graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": ["D"],
    "D": []
}
```

Visualization:

```text
      A
     / \
    B   C
     \ /
      D
```

Advantages of an adjacency list:

- Efficient memory usage
- Fast traversal
- Easy to add new edges
- Ideal for sparse graphs

---

## Time Complexity

| Operation | Time Complexity |
|-----------|----------------:|
| Add Vertex | **O(1)** |
| Add Edge | **O(1)** |
| Traverse Graph (DFS/BFS) | **O(V + E)** |

Where:

- **V** = Number of vertices (nodes)
- **E** = Number of edges (connections)

---

## Real-World Applications of DAGs

Directed Acyclic Graphs are widely used in:

- 📦 Build systems (Gradle, Bazel, Make)
- 🔄 Data engineering pipelines
- 📊 ETL workflows
- ☁️ Cloud orchestration
- 🧠 AI workflow engines
- 📋 Task scheduling systems

They provide a safe way to execute complex workflows while respecting dependencies and preventing infinite execution loops.

## 💻 Basic Python Code (Run in Google Colab)

In the following example, we'll represent a **Directed Acyclic Graph (DAG)** using an adjacency list in Python. We'll create nodes, connect them with directed edges, and traverse the graph to understand how tasks flow through a dependency-based execution pipeline—the same foundational concept used by modern Agentic AI workflow engines.

````
# Representing a basic graph using an ADJANCENCY LIST (Dictionary of nodes and their outgoing neighbors)
# 'graph' is our Adjacency List (adj_list)
graph = {
    "Start": ["AnalyzeInput"],
    "AnalyzeInput": ["ExecuteTool", "RefuseQuery"],
    "ExecuteTool": ["FormatOutput"],
    "RefuseQuery": ["FormatOutput"],
    "FormatOutput": []
}

def traverse_graph(current_node, adj_list):
    """Simple recursive graph traversal"""
    print(f"Visiting Node: {current_node}")
    for neighbor in adj_list.get(current_node, []):
        traverse_graph(neighbor, adj_list)

print("--- Traversing Basic Graph ---")
traverse_graph("Start", graph)
````
