# 📘 Topic 1: Priority Queues & Heaps (Task Schedulers)

---

# Layer 1 — Pure DSA Basics

## 🧠 The Theory

A **Heap** is a specialized **tree-based** data structure that satisfies the **Heap Property**.

The most commonly used variant is the **Min-Heap**, where the smallest element is always stored at the root.

```text
          2
        /   \
       5     7
      / \   / \
     9 10 12 15
```

Every parent node is **less than or equal to** its children.

Because of this property:

- The smallest element can always be accessed instantly.
- Insertion and deletion remain highly efficient.

---

## Why Not Just Sort an Array?

Imagine continuously receiving new tasks.

With a normal array:

1. Insert a task.
2. Sort the entire array.
3. Remove the highest-priority task.

Sorting repeatedly becomes expensive.

Time Complexity:

```text
O(N log N)
```

for every scheduling operation.

---

A **Heap** avoids repeated sorting.

Instead:

- Insert the new task while maintaining the heap property.
- Remove only the root when the next task is needed.

Time Complexity:

| Operation | Complexity |
|-----------|-----------:|
| Peek Minimum | **O(1)** |
| Insert | **O(log N)** |
| Extract Minimum | **O(log N)** |

This makes heaps the standard data structure for **priority scheduling**.

---

## Min-Heap vs Max-Heap

### Min-Heap

```text
        1
      /   \
     3     5
    / \   / \
   7  9  8 10
```

Always removes the **smallest** value first.

---

### Max-Heap

```text
        10
      /    \
     8      7
    / \    / \
   5   4  3   2
```

Always removes the **largest** value first.

Python's built-in `heapq` module implements a **Min-Heap**.

---

## Why Are Heaps So Fast?

Unlike a fully sorted array, a heap only guarantees one thing:

> **The root node is always the highest-priority element.**

The remaining elements only satisfy the local parent-child ordering.

This relaxed ordering allows insertion and deletion to take only:

```text
O(log N)
```

instead of sorting the entire collection after every update.

---

## 💻 Basic Python Code (Run in Google Colab)

In the following example, we'll use Python's built-in **`heapq`** module to create a **Min-Heap**, insert multiple values, and repeatedly extract the smallest element.

This demonstrates the core mechanics of a Priority Queue, which forms the foundation for task scheduling in modern Agentic AI systems.

````
import heapq
import random

# A basic list of numbers we want to manage by priority (smallest first)
raw_tasks = [40, 10, 30, 20, 50]
min_heap = []

# Inserting items into a heap (Heapify / Push)
for task in raw_tasks:
    heapq.heappush(min_heap, task)

print("Min-Heap Structure:", min_heap) # The smallest element (10) is always at index 0

i=1

# Extracting elements in order of priority (smallest first)
while min_heap and i<10:
    next_item = heapq.heappop(min_heap)
    print(f"Extracted highest priority item: {next_item}")
    n=random.randint(1, 100)
    print(f"Inserting new item: {n}")
    heapq.heappush(min_heap,n)
    print("Updated Min-Heap Structure:", min_heap)
    i+=1
````
